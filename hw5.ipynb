```r
# rmultinom standardizes probabilities to 1
# so i just give it the counts
p <- c(10, 68, 112)
```

```r
multinom<-rmultinom(10000,190, p)
```

```r
y <- rep(0,10000)
for (i in 1:10000){y[i] <- (multinom[2, i] + 2 * multinom[3, i]) / 
                  (2 * 190)}
```

```r
hist(y)
```
